The unbiasedness condition requires that the expected value of the estimator equals the true parameter $\theta$

**For Option 1**

$\theta_1= \frac{1}{3} \sum_{n=1}^{3} y_n$

The expected value is:
$E\left[ \frac{1}{3} \sum_{n=1}^{3} y_{n} \right] = \frac{1}{3} \sum_{n=1}^{3} E[y_{n}] = \frac{1}{3} \sum_{n=1}^{3} (\theta + E[\eta_{n}]) = \frac{1}{3} \sum_{n=1}^{3} \theta = \theta$

So, This estimator is unbiased.

**For Option 2**

$\theta_2= \frac{1}{3}\sum_{n=1}^{3} y_{n} + 0.2$

The expected value is:
$E\left[ \frac{1}{3}\sum_{n=1}^{3} y_{n} + 0.2 \right] = E\left[ \frac{1}{3}\sum_{n=1}^{3} y_{n} \right] + 0.2 = \theta + 0.2
$


This estimator is biased because the expected value is $\theta + 0.2$, not $\theta$. 


**For Option 3**

$\theta_3 = \frac{1}{100} \sum_{n=1}^{100} y_n^3$

The expected value is:
$E\left[ \frac{1}{100} \sum_{n=1}^{100} y_n^3 \right] = \frac{1}{100} \sum_{n=1}^{100} E[y_n^3]$

Here, $y_n^3$ includes higher-order terms of the noise $\eta_n$, making this estimator not linear in $y_n$. As a result, the expected value of $y_n^3$ is not simply $\theta$, so this estimator is biased.

**For Option 4**

$\theta_4 = \frac{1}{100} \sum_{n=1}^{100} y_n$

The expected value is:
$E\left[ \frac{1}{100} \sum_{n=1}^{100} y_n \right] = \frac{1}{100} \sum_{n=1}^{100} E[y_n]=  \frac{1}{100} \sum_{n=1}^{100} \theta=\theta$

This estimator is unbiased.

**Options 1 and 4 are unbiased estimators.**

Even though these estimators involve fractions, they are unbiased because these fractions apply equally to both the $\theta$ and noise terms, maintaining the expected value as 
$\theta$.

#### Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
d = 50  # number of datasets
N = 30  # number of data pairs in each dataset
theta_0 = 2  # true value of theta
sigma2 = 64  # variance of the noise

# Prepare to store theta estimates
theta_hats = []

# Generate d datasets and compute LS estimates
for _ in range(d):
    # Generate x values uniformly between 0 and 10
    x = np.random.normal(0, 10, N)
    
    # Generate noise eta ~ N(0, sigma^2)
    eta = np.random.normal(0, np.sqrt(sigma2), N)
    
    # Generate y' values based on y' = theta_0 * x
    y_new = theta_0 * x
    
    # Generate y by adding noise to y'
    y = y_new + eta
    
    # Compute the general LS estimate of theta (theta_hat)
    theta_hat = (x.T @ y) / (x.T @ x)
    theta_hats.append(theta_hat)

# Convert theta_hats to a numpy array for convenience
theta_hats = np.array(theta_hats)

# (c1) Estimate the MSE
MSE = np.mean((theta_hats - theta_0) ** 2)

# (c2) Plot the estimates
plt.figure()
plt.plot(theta_hats, 'bo-', label=r'$\hat{\theta}_i$ values')
plt.axhline(y=theta_0, color='r', linestyle='--', label=r'True $\theta_0 = 2$')
plt.xlabel('Dataset index')
plt.ylabel(r'$\hat{\theta}$')
plt.title(r'LS estimates of $\theta$ over 50 datasets')
plt.legend()
plt.show()

# Print the MSE
print(f"Estimated MSE: {MSE:.5f}")
```

The plot indeed shows variability around the true value 0=2, with most estimates ranging between 1.6 and 2.4. Overall, the variability observed is an expected consequence when dealing with estimates from multiple datasets, each of which carries its own noise and uncertainty. The spread of the estimates (between 1.6 and 2.4) indicates that while the estimates are fairly close to the true value on average, there is still noticeable variance.
